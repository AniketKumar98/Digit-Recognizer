## To import Datasets 

```python
import pandas as pd
```

### Import data

```python
data=pd.read_csv('train.csv')
```

```python
data
```

## Converting it to a matrix.

```python
data=data.as_matrix()
```

#### Below we extract train data from Data Matrix as xtrain which contain pixels value(0 to 255) of all the numbers which is extracted separately in label variable 

```python
xtrain=data[0:35000,1:]
```

```python
xtrain
```

```python
label=data[0:35000,0]
```

```python
label
```

### Import DecisionTreeClassifier which is popular classification technique.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

### Fitting the classifier with data

```python
clf.fit(xtrain,label)
```

## Now Testing Begins

```python
xtest=data[35000:,1:]
```

```python
test_label=data[35000:,0]
```

### Prediction

```python
pred=clf.predict(xtest)
```

### Accuracy

```python
sum(pred==test_label)/70
```
